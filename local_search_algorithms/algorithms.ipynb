```python
# steepest ascent for simple function f(x)

# objective function
def f(x):
    return ((-5/6) * x**4) + (7 * x**3) - ((115/6) * x**2) + (18 * x) + (1/2)


def steepest_ascent(f, iterations):

    cur_x = 0
    stepsize = 1

    for iteration in range(iterations):

        print (f"Iteration {iteration}: f({cur_x}) = {f(cur_x)}")
        stepsize *= 0.99

        left, right = f(cur_x-stepsize), f(cur_x+stepsize)

        if left > right:
            cur_x -= stepsize
        if right > left:
            cur_x += stepsize 
        
    
    print (f"f({cur_x}) = {f(cur_x)}")



steepest_ascent(f, 10)


```

```python
# metropolis test

def probability_min(
    score_new, 
    score_old, 
    temperature):
    return math.exp(
        - ((score_new - score_old) / temperature)
    )

def probability_max(
    score_new, 
    score_old, 
    temperature):
    return math.exp(
        - ((-score_new + score_old) / temperature)
    )

score_new = 90
score_old = 100
temp = 100000

for _ in range(100):
    print(probability_max(score_new, score_old, temp))
    temp *= 0.9
```
